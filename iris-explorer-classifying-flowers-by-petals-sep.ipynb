```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyClassifier
```

```python
# Load the Iris dataset
df = pd.read_csv(filepath_or_buffer='/kaggle/input/iris/Iris.csv')
# Display dataset info (columns, datatypes, and missing values)
print("\nDataset Info:")
df.info()
```

```python
# Check unique species in the 'Species' column
print("\nUnique Species in the dataset before encoding:")
print(df['Species'].unique())
```

```python
# Drop the 'Id' column as it is not useful for the analysis
df.drop(labels=['Id'], axis=1, inplace=True)
print("'Id' column has been dropped from the dataset.")

# Encode the 'Species' column to numeric values using LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
print("'Categories' in 'Species' have been encoded into numeric values.")

# Print the classes in 'Species' after encoding
print("Encoded classes for 'Species':")
print(le.classes_)
```

```python
# Split the data into features (X) and target (y)
X = df.drop(columns='Species')
y = df['Species']
print("Features and target variable separated.")

# Split the data into training and test sets (33% test size)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print("Data has been split into training and test sets (33% test size).")

# Initialize LazyClassifier to fit and evaluate multiple models
print("Initializing LazyClassifier to fit models on the dataset...")
clf = LazyClassifier(
    verbose=0, 
    ignore_warnings=True, 
    custom_metric=None
)

# Fit the models and get predictions
print("Fitting models with LazyClassifier...")
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```
